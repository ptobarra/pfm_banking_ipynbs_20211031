```python
"""
Podemos ver que los cobros se producen en las mismas fechas y por cantidades parecidas
"""
```

```python
position = range(0,6)
color = ['blue','orange','green'] #,'red','purple','brown']

for i,j in zip(position,color):
    years.iloc[:,i].plot(color=j, legend=True)
    pyplot.show()
```
